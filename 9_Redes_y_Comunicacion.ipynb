```python
#Redes y Comunicación
#Crear un red de ocho nodos que muestre una relación
#direccionada, de manera que cada nodo sólo pueda tener
#comunicación directa con otros 2 nodos.

#Importo la libreria networkx que me permite crear y manipular esrtructuras complejas, en este caso los nodos
import networkx as nx

#Creo un grafo dirigido
Nodos = nx.DiGraph()

#Creo mis nodos y les asigno un nombre, este caso una lketra desde la A a la H
Nodos.add_edges_from([("A","B"),("B","C"),("C","D"),("D","E"),("E","F"),("F","G"),("G","H"),("H","A")])

#Input al Usuario donde le muestro mi red completa de nodos, primero la cantidad
print("Redes y Comunicación")
print(" ")
print("Tengo ", len(Nodos.nodes), " nodos en total.")

#Muestro la lista de los nodos y sus conexiones
print(" ")
print("Los nodos son {}".format(list(Nodos.nodes)))
print("Las conexiones entre los nodos son {}".format(list(Nodos.edges)))

#Muestro un detalle de las conexiones de cada nodo
print(" ")
print("Se detallan las conexiones a continuación:")
print(" ")
print("El nodo A trasfiere información hacia el nodo {}".format(list(Nodos.neighbors("A"))), " y recibe información desde el nodo {}".format(list(Nodos.predecessors("A"))))
print(" ")
print("El nodo B trasfiere información hacia el nodo {}".format(list(Nodos.neighbors("B"))), " y recibe información desde el nodo {}".format(list(Nodos.predecessors("B"))))
print(" ")
print("El nodo C trasfiere información hacia el nodo {}".format(list(Nodos.neighbors("C"))), " y recibe información desde el nodo {}".format(list(Nodos.predecessors("C"))))
print(" ")
print("El nodo D trasfiere información hacia el nodo {}".format(list(Nodos.neighbors("D"))), " y recibe información desde el nodo {}".format(list(Nodos.predecessors("D"))))
print(" ")
print("El nodo E trasfiere información hacia el nodo {}".format(list(Nodos.neighbors("E"))), " y recibe información desde el nodo {}".format(list(Nodos.predecessors("E"))))
print(" ")
print("El nodo F trasfiere información hacia el nodo {}".format(list(Nodos.neighbors("F"))), " y recibe información desde el nodo {}".format(list(Nodos.predecessors("F"))))
print(" ")
print("El nodo G trasfiere información hacia el nodo {}".format(list(Nodos.neighbors("G"))), " y recibe información desde el nodo {}".format(list(Nodos.predecessors("G"))))
print(" ")
print("El nodo H trasfiere información hacia el nodo {}".format(list(Nodos.neighbors("H"))), " y recibe información desde el nodo {}".format(list(Nodos.predecessors("H"))))

#Muestro un gráfico de los nodos y sus conexiones
print(" ")
print("El siguiente grafico muestra la distribución de los nodos y sus respectivas conexiones:")
print(" ")
nx.draw(Nodos, with_labels = True, arrows = True)
```
